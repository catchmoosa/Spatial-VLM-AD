```python
def convert_to_rgb(image):
    """Convert image to RGB format if not already in RGB."""
    if image.mode == "RGB":
        return image
    image_rgba = image.convert("RGBA")
    background = Image.new("RGBA", image_rgba.size, (255, 255, 255))
    alpha_composite = Image.alpha_composite(background, image_rgba)
    return alpha_composite.convert("RGB")


def reduce_image_size(image, scale=0.5):
    """Reduce image size by a given scale."""
    original_width, original_height = image.size
    new_width = int(original_width * scale)
    new_height = int(original_height * scale)
    return image.resize((new_width, new_height))


```

```python
def format_data(sample):
    """Format the dataset sample into structured messages."""
    image = sample["images"]
    image = convert_to_rgb(image)
    image = reduce_image_size(image)

    # Extract user and assistant messages
    user_message = sample["messages"][0]["content"]  # Get the content from the user message
    assistant_message = sample["messages"][1]["content"]  # Get the content from the assistant message


    return {
        "messages": [
            {
                "role": "user",
                "content": [
                    {
                        "type": "text",
                        "text": user_message,  # Use the extracted user message
                    },
                    {
                        "type": "image",
                        "image": image,
                    }
                ],
            },
            {
                "role": "assistant",
                "content": [
                    {
                        "type": "text",
                        "text": assistant_message,  # Use the extracted assistant message
                    }
                ],
            },
        ],
    }


# Transform the dataset
converted_dataset = [format_data(sample) for sample in ds]
```

```python
import torch
from unsloth import FastVisionModel
```

```python
model_name = "unsloth/llava-v1.6-mistral-7b-hf"

model, tokenizer = FastVisionModel.from_pretrained(
    model_name = model_name,
    load_in_4bit = True,
    use_gradient_checkpointing = "unsloth",
)
```

```python

from peft import LoraConfig, get_peft_model


# Configure qLoRA
q_lora_config = {
    "bnb_4bit_quant_type": "nf4",          # NormalFloat4 quantization for high-quality 4-bit precision
    "bnb_4bit_compute_dtype": "float16",   # Perform computations in FP16, supported by T4 GPUs
}

# Apply LoRA fine-tuning layers
model = FastVisionModel.get_peft_model(
    model,
    finetune_vision_layers=False,           # Vision layers are frozen
    finetune_language_layers=True,         # Freeze language layers, focus on LoRA
    finetune_attention_modules=True,        # Apply LoRA to attention modules
    finetune_mlp_modules=False,             # Skip MLP layers for qLoRA efficiency
    r=8,                                    # Rank of LoRA matrices
    lora_alpha=16,                          # Scaling factor for LoRA updates
    lora_dropout=0.1,                       # Dropout for better generalization
    bias="none",                            # Typically set to "none" for qLoRA
    random_state=3407,                      # Reproducibility
    use_rslora=False,                       # No RS-LoRA sparsity here
    loftq_config=q_lora_config              # Pass the qLoRA-specific configuration
)

# Verify trainable parameters
model.print_trainable_parameters()
```

```python
FastVisionModel.for_inference(model)         # Enable for inference!

image = ds[5]["images"]
instruction = ds[5]["messages"][0]["content"]

messages = [
    {
        "role": "user",
        "content": [
            {
                "type": "image",
            },
            {
                "type": "text",
                "text": instruction
            },
        ]
    }
]

input_text = tokenizer.apply_chat_template(messages, add_generation_prompt = True)
inputs = tokenizer(
    image,
    input_text,
    add_special_tokens = False,
    return_tensors = "pt",
).to("cuda")
```

```python
from trl import SFTTrainer, SFTConfig
from unsloth import is_bf16_supported
from unsloth.trainer import UnslothVisionDataCollator
```

```python
args = SFTConfig(
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    warmup_steps=5,
    num_train_epochs=1,
    learning_rate=2e-4,
    fp16=not is_bf16_supported(),
    bf16=is_bf16_supported(),
    optim="adamw_8bit",
    weight_decay=0.01,
    lr_scheduler_type="linear",
    seed=3407,
    output_dir="outputs",
    report_to="wandb",
    logging_steps=1,
    logging_strategy="steps",
    save_strategy="no",
    load_best_model_at_end=True,
    save_only_model=False,
    remove_unused_columns=False,
    dataset_text_field="",
    dataset_kwargs={"skip_prepare_dataset": True},
    dataset_num_proc=4,
    max_seq_length=2048,
)
```

```python
FastVisionModel.for_training(model)

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    data_collator = UnslothVisionDataCollator(model, tokenizer),
    train_dataset = converted_dataset,
    args = args,
)
```

```python
# Show current memory stats
gpu_stats = torch.cuda.get_device_properties(0)
start_gpu_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
max_memory = round(gpu_stats.total_memory / 1024 / 1024 / 1024, 3)
print(f"GPU = {gpu_stats.name}. Max memory = {max_memory} GB.")
print(f"{start_gpu_memory} GB of memory reserved.")
```

```python
import os
import json
from transformers import TrainerCallback

# Path to save checkpoints and losses
checkpoint_dir = "/content/drive/My Drive/LlavaCheckpoints"
os.makedirs(checkpoint_dir, exist_ok=True)

# Custom Callback to Save Losses, Model, and Trainer State
class SaveProgressCallback(TrainerCallback):
    def __init__(self, save_path, model, tokenizer, save_every_n_steps=10):
        self.save_path = save_path
        self.losses = []
        self.model = model
        self.tokenizer = tokenizer
        self.save_every_n_steps = save_every_n_steps  # Frequency of saving model/checkpoint

    def on_step_end(self, args, state, control, **kwargs):
        # Ensure log_history has valid entries
        if state.log_history and len(state.log_history) > 0:
            # Safely access the last entry in log_history
            last_log = state.log_history[-1]
            if 'loss' in last_log:
                current_loss = last_log['loss']
                self.losses.append(current_loss)  # Append the current step loss

                # Save losses to a JSON file in Google Drive
                loss_file = os.path.join(self.save_path, "losses.json")
                with open(loss_file, "w") as f:
                    json.dump(self.losses, f)

                print(f"Loss saved for step {state.global_step}: {current_loss}")
            else:
                print(f"Loss key missing at step {state.global_step}, skipping.")

        # Save model, tokenizer, and trainer state every N steps
        if state.global_step % self.save_every_n_steps == 0:
            try:
                # Save model and tokenizer
                self.model.save_pretrained(self.save_path)
                self.tokenizer.save_pretrained(self.save_path)

                # Save trainer state (critical for resumption)
                trainer_state_file = os.path.join(self.save_path, "trainer_state.json")
                state.save_to_json(trainer_state_file)  # Save trainer state

                print(f"Checkpoint saved at step {state.global_step}")
            except Exception as e:
                print(f"Error saving checkpoint at step {state.global_step}: {e}")


# Attach the callback
save_progress_callback = SaveProgressCallback(checkpoint_dir, model=model, tokenizer=tokenizer)
trainer.add_callback(save_progress_callback)
```

```python
# Start Training
trainer_stats = trainer.train()
print(trainer_stats)

# Finalize WandB logging
wandb.finish()
```

```python
# Show final memory and time stats
used_memory = round(torch.cuda.max_memory_reserved() / 1024 / 1024 / 1024, 3)
used_memory_for_lora = round(used_memory - start_gpu_memory, 3)
used_percentage = round(used_memory         /max_memory*100, 3)
lora_percentage = round(used_memory_for_lora/max_memory*100, 3)
print(f"{trainer_stats.metrics['train_runtime']} seconds used for training.")
print(f"{round(trainer_stats.metrics['train_runtime']/60, 2)} minutes used for training.")
print(f"Peak reserved memory = {used_memory} GB.")
print(f"Peak reserved memory for training = {used_memory_for_lora} GB.")
print(f"Peak reserved memory % of max memory = {used_percentage} %.")
print(f"Peak reserved memory for training % of max memory = {lora_percentage} %.")
```
